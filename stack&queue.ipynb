```python
class StackList:
    def __init__(self):
        self.__stack = [] #very imp to make this private so array operation cannot be used for this list
    
    def push(self,data):
        self.__stack.append(data)
        print(f"Pushed {data} into stack")

    def size(self):
        return len(self.__stack)
    
    def isempty(self):
        if (len(self.__stack) == 0):
            return True
        else:
            return False
        
    def top(self):
        if len(self.__stack) == 0:
            print("Stack is empty")
            return None
        # return self.__stack[len(self.__stack) - 1]
        return self.__stack[-1]
    
    def pop(self):
        if len(self.__stack) == 0:
            print("Stack is empty")
            return None
        return self.__stack.pop()
    
st = StackList()
l1 = [1,2,3,4,5,6]
for i in l1:
    st.push(i)
print(f"Size of stack {st.size()}")
print(f"Stack is empty : {st.isempty()}")
print("Top element",st.top())
print("Pop element:",st.pop())
print("Top element",st.top())
    
```

```python
# Stack Using Linked List  to optimize we have to insert through head

class Node():
    def __init__(self,value):
        self.data = value
        self.next = None

class StackUsingLL:
    def __init__(self):
        self.head = None
        self.len = 0
    
    def push(self,data):
        newNode = Node(data)
        self.len += 1 # very imp to maintain size
        if self.head == None:
            self.head = newNode
            return f"Added {data} to the stack"
        
        newNode.next = self.head
        self.head = newNode
        return f"Added {data} to the stack"
    
    def top(self):
        if (self.head == None or self.len == 0):
            return "Stack is Empty,No top element"
        
        return self.head.data
    
    def pop(self):
        if (self.head == None or self.len == 0):
            return "Stack is Empty,Cannot Pop element"
        
        dataAtTop = self.head.data
        self.head = self.head.next
        self.len -= 1
        return dataAtTop
        
    def size(self):
        return self.len
    
    def isempty(self):
        return self.len == 0

st = StackUsingLL()
l1 = [1,2,3,4,5,6]
for i in l1:
    st.push(i)
print(f"Size of stack {st.size()}")
print(f"Stack is empty : {st.isempty()}")
print("Top element",st.top())
print("Pop element:",st.pop())
print("Top element",st.top())


```

```python
# queue using LinkedList

class Node:
    def __init__(self,value):
        self.data =  value
        self.next =  None
class QueueUsingLL:
    def __init__(self):
        self.len = 0
        self.head = None
        self.tail = None
        
    def size(self):
        return self.len
    def isEmpty(self):
        return self.size() == 0
    
    def enqueue(self,data):
        newNode = Node(data)
        self.len += 1

        if self.head == None:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            self.tail = newNode

        return f"Added {data} to the queue"
    
    def front(self):
        if self.isEmpty():
            print("Queue is empty, cannot show front")
            return None
        
        return self.head.data
    
    def dequeue(self):
        if self.isEmpty():
            print("Queue is empty, cannot dequeue")
            return None
        self.len -= 1
        datatobereturn = self.head.data
        self.head = self.head.next
        if (self.head == None): #very imp if len == 1
            self.tail = None
        return datatobereturn

ql = QueueUsingLL()
l1 = [1,2,3,4,5,6]
for i in l1:
    print(ql.enqueue(i))

print(ql.front())
print(ql.dequeue())
print(ql.size())
print(ql.isEmpty())   

            
```

```python
# Monotonic stack
# A monotonic stack is a stack that maintains element in increasing order or decresing order

# Increasing Stack : Keeps elements in increasing order.
# Before pushing new element:
# Pop all elements greater than current

# Monotonic Decreasing Stack
# Keeps elements in decreasing order.
# Before pushing new element:
# Pop all elements smaller than current



def next_greater_element(arr):
    n= len(arr)
    result = [-1] * n
    stack = []

    for i in range(n-1,-1,-1):
        while(stack and stack[-1] <= arr[i]):
            stack.pop()
        
        if stack:
            result[i] = stack[-1]
        stack.append(arr[i])
    return result

a = [4,5,2,25]
next_greater_element(a)
```

```python
# Example:
# Input: arr = [10, 20, 30, 40]
# Output: [-1, -1, -1, -1]
# Input: arr = [1, 3, 2, 4]
# Output: [-1, 2, -1, -1]
# Input: arr = [4, 5, 2, 10, 8]
# Output: [2, 2, -1, 8, -1]
def next_smaller_element(arr):
    n = len(arr)
    result = [-1] * n
    stack = []
    for i in range(n-1,-1,-1):
        while(stack and stack[-1] >= arr[i]):
            stack.pop()

        if stack:
            result[i] = stack[-1]
        
        stack.append(arr[i])
    return result

input = [10,20,30,40]
ins = [1, 3, 2, 4]
print(next_smaller_element(input))
print(next_smaller_element(ins))
    
```

```python
# Valid Parenthesis
# Problem statement You're given a string 'S' consisting of "{", "}", "(", ")", "[" and "]" . Return true if the given string 'S' is balanced, else return false. '
# 'For example: 'S' = "{}()". There is always an opening brace before a closing brace i.e. '{' before '}', '(' before '). So the 'S' is Balanced.
# brute force On2
def validPare(S):
    prev_len = -1
    while (prev_len != len(S)):
        prev_len = len(S)

        S = S.replace("()","")
        S = S.replace("{}","")
        S = S.replace("[]","")

        return len(S) == 0
    

St = '{}()'
validPare(St)
```

```python
def validParenthesis(s):
    stack = []
    for bracket in s:
        if bracket == '(' or bracket == '{' or bracket == '[':
            stack.append(bracket)
        else:
            if len(stack) == 0:
                return False
            ch = stack.pop()
            if (bracket == ')' and ch == '(') or (bracket == '}' and ch == '{') or (bracket == ']' and ch == '['):
                continue
            else:
                return False
    return len(stack) == 0

St = '()'
validParenthesis(St)
```
