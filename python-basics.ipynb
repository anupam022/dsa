```python
# count word frequency from file
def count_word_frequency(file_path):
    try:
        with open(file_path, 'r') as file:
            text = file.read()
            words = text.split()
            frequency = {}
            for word in words:
                word = word.lower().strip('.,!?";:()[]{}')  # Normalize the word
                if word in frequency:
                    frequency[word] += 1
                else:
                    frequency[word] = 1
            return frequency
    except FileNotFoundError:
        print("File not found.")
        return {}
    
result = count_word_frequency('sample.txt')
print(result)    
```

```python
# Square of side 'N'
# Input: 3
# Output: ['***', '***', '***']
 
# Input: 5
# Output: ['*****', '*****', '*****', '*****', '*****']

def genrate_square(n):
    return list(map(lambda x:'*' * n,range(n)))
print(genrate_square(5))
```

```python
# Hollow Square of side 'N'
# Input: 3
# Output: ['***', '* *', '***']
 
# Input: 5
# Output: ['*****', '*   *', '*   *', '*   *', '*****']

def hollow_square(n):
    return list(map(lambda i: '*' * n if i == 0 or i == n - 1 else '*' + ' ' * (n - 2) + '*', range(n)))
print(hollow_square(7))
```

```python
# Right angled triangle
# Input: 3
# Output: ['*', '**', '***']
 
# Input: 5
# Output: ['*', '**', '***', '****', '*****']

def right_angled_triangle(n):
    return list(map(lambda x: '*' * (x + 1), range(n)))
print(right_angled_triangle(5))
```

```python
# Inverted Right Angled Triangle
# Input: 3
# Output: ['***', '**', '*']
 
# Input: 5
# Output: ['*****', '****', '***', '**', '*']

def inverted_right_angled_triangle(n):
    return list(map(lambda x: '*' * (n - x), range(n)))
print(inverted_right_angled_triangle(5))
```

```python
# Pyramid Pattern
#  The pyramid is centered, with 1 star in the first row, 3 stars in the second row,
#  and so on, increasing by 2 stars per row until the base row has 2n - 1 stars.
# Input: 3
# Output: ['  *  ', ' *** ', '*****']
 
# Input: 5
# Output: ['    *    ', '   ***   ', '  *****  ', ' ******* ', '*********']

def pyramid_pattern(n):
    return list(map(lambda i: ' ' * (n - i - 1) + '*' * (2 * i + 1) + ' ' * (n - i - 1), range(n)))
print(pyramid_pattern(5))
```

```python
# Inverted Pyramid Pattern
# Input: 3
# Output: ['*****', ' *** ', '  *  ']
 
# Input: 5
# Output: ['*********', ' ******* ', '  *****  ', '   ***   ', '    *    ']
def inverted_pyramid(n):
    return list(map(lambda i: ' ' * i + '*' * (2 * (n - i) - 1) + ' ' * i, range(n)))
print(inverted_pyramid(3))
```

```python
# Maximum difference between two consecutive elements in a list.
# Input: lst = [1, 7, 3, 10, 5]
# Output: 7

# The maximum difference is between 3 and 10 (i.e., |3 - 10| = 7).

# Input: lst = [10, 11, 15, 3]
# Output: 12

# The maximum difference is between 15 and 3 (i.e., |15 - 3| = 12).

def maxdiffcon(lst):
    diff = []
    for i in range(len(lst) - 1):
        diff.append(abs(lst[i] - lst[i+1]))
    
    return max(diff) if len(lst) > 2 else diff[0] if len(diff) > 0 else 0
    
lst = [1, 7, 3, 10, 5]
print(maxdiffcon(lst))  
```

```python
# Rotate a List
# Input: lst = [1, 2, 3, 4, 5], k = 2
# Output: [4, 5, 1, 2, 3]

# Input: lst = [10, 20, 30, 40, 50], k = 3
# Output: [30, 40, 50, 10, 20]

def rotate_list(lst, k):
    if not lst:
        return []
    result = []
    n = len(lst)
    print("length",n)
    k = k%n
    print("Value of k",k)
    for i in range(n-k,n):
        result.append(lst[i])
    
    for i in range(0,n-k):
        result.append(lst[i])
    return result
          
# lst = [1, 2, 3, 4, 5] 
lst = [10, 20, 30, 40, 50]
k=3
print(rotate_list(lst,k))
            
```

```python
# Words Frequency in a Sentence
# Input: "hello world hello"
# Output: {'hello': 2, 'world': 1}

# Input: "the quick brown fox jumps over the lazy dog"
# Output: {'the': 2, 'quick': 1, 'brown': 1, 'fox': 1, 'jumps': 1, 'over': 1, 'lazy': 1, 'dog': 1}

def count_word_frequency(sentence):
    words = sentence.split()
    frequency = {}
    for word in words:
        word = word.lower()  # Normalize the word to lowercase
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency

sentence = "hello world hello"
count_word_frequency(sentence)
```

```python
# Linear Search arr,target  --> Return the target index if present else return -1
#Run a loop and find target

def linear_search(arr,target):
    size = len(arr)-1
    for index in range(0,size):
        if(arr[index] == target):
            return index
    return -1
    
my_list = [10,20,40,30,4]
target = 30

result = linear_search(my_list,target)
print(result)
```

Binary Search Algorithm

```python
# Binary Search can only be applied to sorted array   arr = [10,23,35,45,50,70,85] target = 50
# I will consider full array/list start = 0 end = n-1  (n is whole array length)
# See the middle element mid = (end-start)//2   1st iteration => mid = 3 arr[mid] = 45 <50 
# We compare element at mid with target
# 2nd Iteration I will start with greater than mid index of previous one because mid previous value was smaller than target 
# 2nd iteration mid ==> 6+4//2 = 5 arr[mid] = 70 since 70 > 50 then we will move end to middle -1 which will be start and end at same position mid = 4+4//2 = 4 arr[mid] =50 == target
def binary_Search(arr,target):
    size = len(arr)
    start = 0
    end = size-1
    while(start <= end):
        mid = (start + end)//2
        if arr[mid] == target:
            return [mid,arr[mid]]
        elif (arr[mid] > target):
            end = mid -1
        elif (arr[mid] < target):
            start = mid + 1
        else:
            return -1
    

my_list = [10,23,35,45,50,70,85]
target = 85
result = binary_Search(my_list,target)
print(result)
```

Bubble Sort Algorithm 

```python
# arr = [12,25,11,34,90,22] 
# Start with the first element of the list.
# Compare the current element with the next element.
# If the current element is greater, swap them.
# Move to the next pair and repeat step 2 and 3 for the entire list.
# After each pass, the largest element moves to the end.
# Repeat the process for the remaining unsorted part of the list.
# Stop when no swaps are needed in a pass (the list is sorted).
# 
def bubble_sort(arr):
    n = len(arr)
    for passes in range(n):
        for i in range(0,n-1-passes):
            if(arr[i] > arr[i+1]):
                arr[i],arr[i+1] = arr[i+1],arr[i]
    return arr
     
random_list = [12,25,11,34,90,22]
result = bubble_sort(random_list)
print("Sorted List : ",result)
```

Selection Sort

```python
# arr = [12,25,11,34,90,22]
# In each pass we see the array and select the minimum we swap to the right position
# Start at the first element of the list.
# Find the minimum element in the unsorted part of the list.
# Swap the minimum element with the first unsorted element.
# Move the boundary of the sorted part one step forward.
# Repeat steps 2–4 for the remaining unsorted part until the list is sorted.
def selection_sort(arr):
    n=len(arr)
    for i in range(n-1):
        min_idx = i
        for j in range(i+1,n):
            if (arr[j]<arr[min_idx]):
                min_idx = j
        arr[i],arr[min_idx] = arr[min_idx],arr[i]
    return arr    


unsorted_list = [12,25,11,34,90,22]
result = selection_sort(unsorted_list)
print("Sorted List : ",result)
```

Insertion Sort

```python
# Start from the second element (index 1), considering the first element as sorted.
# Pick the current element and compare it with elements in the sorted part (to its left).
# Shift elements in the sorted part that are greater than the current element one position to the right.
# Insert the current element into its correct position in the sorted part.
# Repeat steps 2–4 for all elements until the list is sorted.
# arr = [12, 25, 11, 34, 90, 22]
def insertion_sort(arr):
    n = len(arr)
    for current in range(1,n):
        currentCard = arr[current]
        correctposition = current - 1
        while correctposition >=0 and arr[correctposition] > currentCard:
            arr[correctposition + 1] = arr[correctposition]
            correctposition -= 1
        arr[correctposition + 1] = currentCard
    
    return arr

arr = [12,25,11,34,90,22]
print(insertion_sort(arr))
```

```python
# Count negative numbers in a sorted matrix
# Parameters:
# grid (List[List[int]]): A 2D matrix with dimensions m x n, where each row and each column is sorted in non-increasing order.
# Return Values:
# Integer: The count of negative numbers in the matrix.
# Input: grid = [[4, 3, 2, 1], [3, 2, 1, -1], [1, 1, -1, -2], [-1, -1, -2, -3]] 
# Output: 7 
# Explanation: There are 7 negative numbers in the matrix.
 
# Input: grid = [[3, 2], [1, 0]] 
# Output: 0 
# Explanation: There are no negative numbers in the matrix.
# Brute force approach
"""def countNegatives(grid):
    count = 0
    for row in grid:
        for col in row:
            if col < 0:
                count +=1
    return count"""

def countNegatives(grid):
    total = 0
    cols = len(grid[0])
    for row in grid :
        start, end = 0, cols -1 
        while start <= end:
            mid = (start  + end)//2
            if row[mid] <0:
                end = mid -1
            else:
                start = mid + 1
        total += (cols-start)
    return total
                
    
input = [[4, 3, 2, 1], [3, 2, 1, -1], [1, 1, -1, -2], [-1, -1, -2, -3]]
print("Count negative : ",countNegatives(input))
```

```python
# Find smallest letter greater than target
"""
    Return the smallest character in letters that is lexicographically greater than target.

    Parameters:
    letters (List[char]): Sorted array of characters.
    target (char): The target character.

    Returns:
    char: The smallest character greater than target, or the first character if no such character exists.
    """
# Input:
# letters = ['c', 'f', 'j']
# target = 'k'
# Output: 'c'
 
# Input:
# letters = ['c', 'f', 'j']
# target = 'c'
# Output: 'f'
 
# Input:
# letters = ['c', 'f', 'j']
# target = 'a'
# Output: 'c'

def next_greatest_letter(letters, target):
    size = len(letters)
    start,end = 0, size - 1
    result = letters[0]
    while(start<=end):
        mid = (start + end)//2
        if letters[mid] > target:
            result = letters[mid]
            end = mid -1
        else:
            start = mid + 1
    return result
letters = ['c', 'f', 'j']
target = 'c'
print("Smallest Letter greater than target :",next_greatest_letter(letters,target))
            
```

```python
# Find First and Last Position of Element in Sorted Array
# Input: nums = [5, 7, 7, 8, 8, 10], target = 8 
# Output: [3, 4] 
# Explanation: The target 8 appears from index 3 to index 4.
 
 
# Input: nums = [5, 7, 7, 8, 8, 10], target = 6 
# Output: [-1, -1] 
# Explanation: The target 6 is not found in the array.
def searchRange(nums, target):
    def find_bound(is_first):
        left , right = 0 , len(nums) - 1
        bound = -1
        while left <= right:
            mid = (left + right )//2
            if nums[mid] == target:
                bound = mid
                if is_first:
                    right = mid - 1 #search left
                else:
                    left = mid + 1
            elif nums[mid] <target:
                left = mid + 1
            else:
                right = mid -1
        return bound 
    first = find_bound(True)
    last = find_bound(False)
    return [first,last]   
    
        
input = [5, 7, 7, 8, 8, 10]
target = 8
print("First and Last Position : ",searchRange(input,target))
            

```
