```python
# Take input of Linked List
# return head to a newly created LL
# Optimized approach
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None

def print_LLs(head):
    temp = head
    while(temp != None):
        print(temp.data,end="->")
        temp = temp.next
    # print()
    return

def take_optimized_input():
    value = int(input("Enter the value of Node :- "))
    head = None
    tail = None
    while(value !=-1):
        newNode = Node(value)
        if(head == None):
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
        
        value = int(input("Enter the value of Node :- "))
    return head
    
newHead = take_optimized_input()
print_LLs(newHead)
```

```python
# Length of LL in normal way
def length_LL(head):
    temp = head
    ans = 0
    while(temp != None):
        temp = temp.next
        ans += 1
    return ans

headLL = take_optimized_input()
length = length_LL(headLL)
print(length)
```

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None

def tookinput():
    val = int(input("Enter the value : "))
    head = None
    tail = None
    while(val != -1):
        newNode = Node(val)
        if (head == None):
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
        val = int(input("Enter the value : "))
    return head

def printL(head):
    temp = head
    while(temp != None):
        print(temp.data,end="->")
        temp = temp.next

    return

def insertAtHead(head,data):
    newNode = Node(data)
    if data == -1:
        return head
    newNode.next = head
    head = newNode
    return head

def inserAtTail(head,data):
    newNode = Node(data)
    if data == -1:
        return head
    if (head is None):
        return newNode
    temp = head
    while(temp.next != None):
        temp = temp.next
    temp.next = newNode

    return head

def insertatTail_Recur(head,data):
    newNode = Node(data)
    if head == None:
        return newNode
  
    head.next = insertatTail_Recur(head.next,data)

    return head

def insertAtIndex(head,data,index):
    newNode = Node(data)
    if index == 0:
        newNode.next = head
        head = newNode
        return head
    
    temp = head
    count = 0
    while (temp != None and count < index -1):
        temp = temp.next
        count += 1 
    if (temp is None):
        print("Index out of bounds")
        return head
    newNode.next = temp.next
    temp.next = newNode
    return head

def deleteNodeLLHEAD(head):
    if head == None:   #My LL is empty
        return None
    newHead = head.next
    return newHead

def is_tail_Node(node):  #supporter function for while(temp.next.next)
    if(node == None):
        return True
    if(node.next == None):
        return True
    return False

def deleteNodeTail(head):
    if (head == None or head.next == None):
        return None
    
    temp = head
    while(temp.next.next != None):
        temp = temp.next
    temp.next = None
    return head

def deleteNodeAtIndex(head,index):
    if head == None:
        print("LL is empty")
        return head
    if (index==0):
        return  head.next
    temp =head
    count = 0

    while(temp != None and count < index - 1 ):
        temp = temp.next
        count += 1

    if temp == None:
        print("Out of bounds")
        return head
    
    nodetobedeleted = temp.next
    nodeafterdeleted = nodetobedeleted.next

    # temp.next = temp.next.next
    temp.next = nodeafterdeleted
    return head

def deletenodebyvalue(head,value):
    if head == None:
        print("LL is empty")
        return None
    if(head.data == value):
        return head.next  #boundary case when head is value
    
    temp = head
    while(temp.next is not None and temp.next.data != value):
        temp = temp.next

    if temp.next is None:
        print("Value not found")
        return head
    
    nodetobedeleted = temp.next
    nodeafterdeleted = nodetobedeleted.next

    temp.next = nodeafterdeleted
    # temp.next = temp.next.next
    return head


def searchNodeValue(head,value):
    if head == None:
        print("LL is empty")
        return None
    temp = head
    index = 0
    while(temp != None):
        if(temp.data == value):
            return index
        temp = temp.next
        index += 1
    
    return "Not found"
    

 
head = tookinput()
print("Linked List : ")
printL(head)
print("\n")
data = int(input("Enter the head value : "))
head = insertAtHead(head,data)
print("After inserting head \n")
printL(head)

print("\n")
data = int(input("Enter the tail value : "))
head = inserAtTail(head,data)
print("After inserting tail \n")
printL(head)

print("\n")
data = int(input("Enter the value to add at given index : "))
index = int(input("Enter the index : "))
print(index)
head = insertAtIndex(head,data,index)
print("After inserting at Index \n")
printL(head)

print("\n")
head = deleteNodeLLHEAD(head)
print("After deleting head node \n")
printL(head)

print("\n")
head = deleteNodeTail(head)
print("After deleting tail node \n")
printL(head)

print("\n")
index = int(input("Enter the index to delete: "))
print(index)
head = deleteNodeAtIndex(head,index)
print("After deleting at Index \n")
printL(head)

print("\n Deleting node by value")
val = int(input("Enter the value to delete: "))
print(val)
head = deletenodebyvalue(head,val)
print("After deleting the value \n")
printL(head)

print("\n Searching for a value in the linked list")
val = int(input("Enter the value to search: "))
print(val)
index = searchNodeValue(head,val)
print("The value is at index: ",index)


```
