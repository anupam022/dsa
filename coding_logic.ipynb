```python
# Level 1: Simple Conditions (Getting started) 
# 1. Take a number and print whether it’s positive, negative, or zero. 
# 2. Check if a number is even or odd. 
# 3. Check if a number is divisible by 5. 
# 4. Check if a number is divisible by both 3 and 5. 
# 5. Check if a given year is a leap year. 
# 6. Take two numbers and print the larger one. 
# 7. Take three numbers and print the largest. 
# 8. Take a temperature value and print “Cold”, “Warm”, or “Hot” using range conditions. 
# 9. Take a character and check if it’s a vowel or consonant. 


def printNo():
    value = int(input("Enter the number :"))
    if value == 0:
        return "zero"
    return "Positive" if value > 0 else "Negative"
printNo()
```

```python
def evenOdd(val):
    return "Even" if val % 2 == 0 else "Odd"

l = [1,2,3,4,5]
for i in l:
    print(evenOdd(i))
```

```python
def div5(val):
    return "Divisible by 5" if val % 5 == 0 else "Not divisible"

l = [1,25,3,4,5]
for i in l:
    print(div5(i))
```

```python
def div35(val):
    return "Divisible by 3 & 5" if val %3 == 0 and val % 5 == 0 else "Not divisible"


l = [45,24,48,4,5]
for i in l:
    print(div35(i))

```

```python
# every year that is exactly divisible by 4 is a leap year, except for years that are exactly divisible by
#  100; the centurial years that are exactly divisible by 400 are still leap years. For example, the years 1700, 1800, and 1900 are not leap years, but the years 1600 and 2000 are.
def leapyear(year):
    if (year % 4 ==0 and year % 100 != 0) or (year % 400 == 0):
        return "Leap year"
    return "Not leap year"

leapyear(2012)
    
```

```python
def primeNo(no):
    if no < 2:
        return "Not prime"
    for i in range(2, no):
        if no % i == 0:
            return "Not prime"
    return "prime"
        

primeNo(17)
```

```python
# Take two numbers and print the larger one.

def larger(a,b):
    return a if a>b else b

larger(7,5)
```

```python
# take three number print the largest
# a = 5
# b = 6
# c = 4

def largerIn3(a,b,c):
    if a>b and a>c:
        return a
    if b>a and b>c:
        return b
    if c>a and c>b:
        return c
    
largerIn3(4,7,9)
```

```python
# Take a temperature value and print “Cold”, “Warm”, or “Hot” using range conditions. 
# Interpret temperature in Celsius.

# define ranges: cold < 10, warm 10-25, hot > 25 (example thresholds)

def classify_temp(celsius):
    if celsius <= 0:
        return "Cold"
    elif celsius <= 25:
        return "Warm"
    else:
        return "Hot"

# examples
for t in [-5, 0, 10, 25, 30]:
    print(t, classify_temp(t))
```

```python
# Take a character and check if it’s a vowel or consonant. 
def charcheck(char):
    if char in ['a','e','i','o','u']:
        return "Vowel"
    return "consonant"
for i in "Saundrya":
    print(i,charcheck(i))
```

```python
# Take a character and check whether it’s uppercase, lowercase, a digit, or a special character. 
def characterLUDS(char):
    if char.isupper():
        return "Uppercase"
    elif char.islower():
        return "Lowercase"
    elif char.isdigit():
        return "Digit"
    else:
        return "Special Character"
    
for i in ['abc','Abc',"%%%",'ABC','123']:
    print(i,characterLUDS(i))
```
