```python
# pattern 1
for i in range(0,4):
    for j in range(0,4):
        print("*", end = " ")
    print()
```

```python
# pattern 2 Increasing pattern
for i in range(5):
    for j in range(i+1):
        print("*", end = " ")
    print()
```

```python
# pattern 3 decreasing pattern
for i in range(5):
    for j in range(i,5):
        print("*", end = " ")
    print()
```

```python
# pattern 4
n = 5
for i in range(n):
    for j in range(i,n):
        print(" ",end = " ")
    for j in range(i+1):
        print("*",end=" ")

    print()
```

```python
# pattern 5
n = 5
for i in range(n):
    for j in range(i+1):
        print(" ",end = " ")
    for j in range(i,n):
        print("*",end=" ")

    print()
```

```python
# pattern 6
n = 5
for i in range(n):
    for j in range(i,n):
        print(" ",end = " ")
    for j in range(i):
        print("*",end=" ")
    for j in range(i+1):
        print("*",end=" ")
    print()
```

```python
# pattern 7
n=5
for i in range(n):
    for j in range(i+1):
        print(" ",end = " ")
    for j in range(i,n):
        print("*",end = " ")
    for j in range(i,n-1):
        print("*",end = " ")
    print()
```

```python
# Pattern 8 diamond pattern
n=5
for i in range(n-1):
    for j in range(i,n):
        print(" ",end = " ")
    for j in range(i):
        print("*",end=" ")
    for j in range(i+1):
        print("*",end=" ")
    print()
for i in range(n):
    for j in range(i+1):
        print(" ",end = " ")
    for j in range(i,n):
        print("*",end = " ")
    for j in range(i,n-1):
        print("*",end = " ")
    print()
```
