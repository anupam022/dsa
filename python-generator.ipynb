```python
# In Python, an iterator is an object that allows you to iterate (loop) through a sequence — such as a list, tuple, string, or even a custom object — one element at a time.
my_list = [1,2,3,4,5]
for i in my_list:
    print(i,'')
```

```python
# Iterator
iterator = iter(my_list)
print(type(iterator))
```

```python
print(iterator)
```

```python
# Iterate through all element 
iterator = iter(my_list)
try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator.")
```

```python
# Iterators let you access elements one at a time instead of loading everything into memory.
nums = [i for i in range(1_000_000)]  # uses a lot of memory
```

```python
nums = iter(range(1_000_000))  # uses very little memory
print(next(nums))  # only one item in memory at a time
```

# Generators
Generators are a simply way to create iterators. They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

```python
def square(n):
    for i in range(3):
        yield i**2
square(3)

for i in square(3):
    print(i)
```

```python
def my_gen():
    yield 1
    yield 2
    yield 3

gen = my_gen()

for val in gen:
    print(val)
```

```python
# # Reading large files
def read_large_file(file_path):
    with open(file_path,'r') as file:
        for line in file:
            yield line
file_path = 'large_file.txt'
for line in read_large_file(file_path):
    print(line.strip())
```

# Decorators 
Decorators are powerful and flexible feature in Python that allows you to modify the behaviour of a function or class method. They are commonly used to add functionality to functions or methods without modifying their actual code.

```python
# Function copy
def welcome():
    return "Welcome to learning of Decorators"
welcome()
```

```python
wel = welcome
print(wel())

del welcome
print(wel())
```

```python
# Closures Function inside function

def main_welcome(msg):
    
    def sub_welcome():
        print("Welcome to advance closure function")
        print(msg)
        print("Please learn concepts")
    return sub_welcome()

main_welcome("Hello")
```

```python
def main_welcome(func):
    
    def sub_welcome():
        print("Welcome to advance closure function")
        func("Welcome and check")
        print("Please learn concepts")
    return sub_welcome()

main_welcome(print)
```

```python
### Decorators
def main_welcome(func):
    
    def sub_welcome():
        print("Welcome to advance closure function")
        func()
        print("Please learn concepts")
    return sub_welcome()

main_welcome(course_intro)
```

```python
def course_intro():
    print("This is an advanced python course")

course_intro()
```

```python
@main_welcome
def course_intro():
    print("This is an advanced python course")
```

```python
### Decorator ex 2
def my_decor(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper
```

```python
@my_decor
def say_hello():
    print("Hello!")
say_hello()
```

```python
##Decorator with arguments
def repeat(n):
    def decorator(func):
        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        return wrapper
    return decorator

@repeat(4)
def say_hello():
    print("Hello!")

say_hello()
```
